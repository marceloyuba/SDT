FUNCIÓN PARA IMPRIMIR LAS URL Y LAS VARIABLES

```python
def load_and_print_concatenation(start_year, end_year, letter_type, url_replace):
    # Lista para almacenar los nombres de los dataframes por año
    df_names_by_year = []

    # Iterar sobre el rango de años
    for year in range(start_year, end_year + 1):
        # Lista para almacenar los nombres de los dataframes para este año
        df_names_this_year = []

        # Iterar sobre los 12 meses
        for month in range(1, 13):
            # Formatear el número del mes con dos dígitos
            month_str = f"{month:02d}"
            # Construir el sufijo del año
            year_suffix = str(year)[-2:]
            # Construir el nombre de la URL y la variable
            url_var_name = f"url_{month}_{year_suffix}"
            df_var_name = f"{letter_type}_{month}_{year_suffix}"
            # Construir la URL para el mes actual
            url = f"https://d37ci6vzurychx.cloudfront.net/trip-data/{url_replace}_tripdata_{year}-{month_str}.parquet"
            # Imprimir la URL
            print(f"{url_var_name} = \"{url}\"")
            # Imprimir la lectura del dataframe
            print(f"{df_var_name} = pd.read_parquet({url_var_name})")
            # Agregar el nombre del dataframe a la lista de este año
            df_names_this_year.append(df_var_name)

        # Agregar la lista de nombres de dataframes de este año a la lista principal
        df_names_by_year.append(df_names_this_year)

    # Imprimir cómo se concatenarían los dataframes por año
    for year, df_names_this_year in zip(range(start_year, end_year + 1), df_names_by_year):
        df_year_var_name = f"df_{year}_{letter_type}"
        print(f"{df_year_var_name} = pd.concat([{', '.join(df_name for df_name in df_names_this_year)}], ignore_index=True)")

#imprimimos las variables 
load_and_print_concatenation(2018, 2023, 'g', 'green')
```

FUNCION PARA VALIDAR SI EL AÑO ES CORRECTO DE LO CONTRARIO ELIMINARLO

```python
def ajuste_anio(df, year):
   
    # Filtrar las filas donde las fechas no coinciden con el año del df
    df = df[(df['lpep_pickup_datetime'].dt.year == year) & (df['lpep_dropoff_datetime'].dt.year == year)]

    # Reestablecemos los indices
    df.reset_index(drop=True, inplace=True)
    return df
# Aplicar la función a cada DataFrame

df_2018_g = ajuste_anio(df_2018_g,2018)
df_2019_g = ajuste_anio(df_2019_g,2019)
df_2020_g = ajuste_anio(df_2020_g,2020)
df_2021_g = ajuste_anio(df_2021_g,2021)
df_2022_g = ajuste_anio(df_2022_g,2022)
df_2023_g = ajuste_anio(df_2023_g,2023)
```

FUNCION PARA EXTRAER LAS FECHAS Y AGREGAR LAS COLUMNAS PARA AÑO, MES Y DIA_INICIO, DIA_FIN Y HORA _INICIO  HORA_FIN

```python
def separar_columnas_fecha(df, columna_partida, columna_destino):
   

    # Separar las columnas de fecha y hora en componentes individuales
    df['anio'] = df[columna_partida].dt.year
    df['mes'] = df[columna_partida].dt.month
    df['dia_inicio'] = df[columna_partida].dt.day
    df['hora_inicio'] = df[columna_partida].dt.time
    
    df['dia_fin'] = df[columna_partida].dt.day
    df['hora_fin'] = df[columna_destino].dt.time
    #df = df.reindex(columns=column_order)
    return df
```

```python
#implementamos la función anterior
df_green_trip_data_2018_2021 = separar_columnas_fecha(df_green_trip_data_2018_2021,'lpep_pickup_datetime','lpep_dropoff_datetime')
df_green_trip_data_2022_2023 = separar_columnas_fecha(df_green_trip_data_2022_2023,'lpep_pickup_datetime', 'lpep_dropoff_datetime')
```

```python
#funcion que agrega columna de tiempo de viaje y a la vez se eliminan las columnas 'lpep_dropoff_datetime' y 'lpep_pickup_datetime'
def calcular_tiempo_total_de_viaje(df):
    # Calcula la diferencia entre la columna de dropoff y pickup para obtener el tiempo de viaje
    df['tiempo'] = df['lpep_dropoff_datetime'] - df['lpep_pickup_datetime']
    
    # Extrae los componentes de tiempo: horas, minutos y segundos
    df['horas'] = df['tiempo'].dt.components['hours']
    df['minutos'] = df['tiempo'].dt.components['minutes']
    df['segundos'] = df['tiempo'].dt.components['seconds']
    # Combina las columnas de horas, minutos y segundos en una sola columna
    df['total_tiempo'] = df['horas'].astype(str).str.zfill(2) + ':' + df['minutos'].astype(str).str.zfill(2) + ':' + df['segundos'].astype(str).str.zfill(2)
    df.drop(columns=['tiempo','horas','minutos', 'segundos','lpep_dropoff_datetime','lpep_pickup_datetime'],inplace= True)
    return df
```

```python
#llamamos a la función calcular_tiempo_de_viaje
df_green_trip_data_2018_2021 = calcular_tiempo_total_de_viaje(df_green_trip_data_2018_2021)
df_green_trip_data_2022_2023 = calcular_tiempo_total_de_viaje(df_green_trip_data_2022_2023)
```

```python
def asignar_color(df_viajes):
    def asignar_color_por_location(location):
        if location not in [264, 265]:
            return 'Green'
        else:
            return 'ND'

    df_viajes['tipo_color'] = df_viajes['PULocationID'].apply(asignar_color_por_location)
    return df_viajes
```
